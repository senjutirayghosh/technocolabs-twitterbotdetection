```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

print("Setup Complete")
```

```python
#merging all positive datasets
df1 = pd.read_csv("IRAhandle_tweets_1[1].csv")
df2 = pd.read_csv("IRAhandle_tweets_2[1].csv")
df3 = pd.read_csv("IRAhandle_tweets_3[1].csv")
df4 = pd.read_csv("IRAhandle_tweets_4[1].csv")
df5 = pd.read_csv("IRAhandle_tweets_5[1].csv")
df6 = pd.read_csv("IRAhandle_tweets_6[1].csv")
df7 = pd.read_csv("IRAhandle_tweets_7[1].csv")
df8 = pd.read_csv("IRAhandle_tweets_8[1].csv")
df9 = pd.read_csv("IRAhandle_tweets_9[1].csv")

pos_tweets_full = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], axis = 0)
pos_tweets_full
```

```python
#removing all non-English tweets
pos_tweets = pos_tweets_full.loc[pos_tweets_full.language == 'English']
pos_tweets = pos_tweets.sample(frac = 1)
pos_tweets
```

```python
#extracting the content of the final positive dataset
pos_tweets = pd.DataFrame(pos_tweets, columns = ['content'])

#labelling them as 1 (bot)
pos_tweets['label'] = 1

pos_tweets = pos_tweets.iloc[:10000]
pos_tweets
```

```python
#converting negative dataset from its .txt file type to .csv format
neg_tweets = pd.read_csv("tweets-2016-10000-textonly.txt", delimiter = '\t', header = None)
neg_tweets.columns = ['content']
neg_tweets.to_csv("tweets-2016-10000-textonly.csv", index = None)

#labelling negative tweets as 0 (non-bot)
neg_tweets['label'] = 0

neg_tweets
```

```python
#merging the newly formed datasets containing positive and negative tweets
df_final = pd.concat([pos_tweets, neg_tweets], axis = 0)
df_final
```

```python
#shuffling the rows of the final merged dataset containing all positive and negative datasets
df_final = df_final.sample(frac = 1).reset_index(drop = True)
df_final
```

```python
#Saving the updated dataframe
df_final.to_csv("Final dataset.csv")
df_final
```
